```python
# Import library yang dibutuhkan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('max_columns', 120)
```

```python
# Membaca dataset loan 2007-2014
df = pd.read_csv('loan_data_2007_2014.csv')
# Melihat 5 baris pertama
df.head()
```

```python
# Drop kolom pertama karena hanya merupakan index
df = df.drop(df.columns[0], axis=1)
df.head()
```

```python
# Memvisualisasi dataset menggunakan heatmap untuk melihat berapa banyak kolom yang mengandung missing value atau nilai kosong
plt.figure(figsize=(20,10))
sns.heatmap(df.isnull(),yticklabels=False,cbar=False, cmap='viridis')
```

```python
# Membuang kolom yang mengandung missing value sebanyak 50% data atau lebih
df = df[df.columns[df.isnull().sum()<len(df)/2]]
plt.figure(figsize=(20,10))
sns.heatmap(df.isnull(),yticklabels=False,cbar=False, cmap='viridis')
```

```python
'''Setelah membaca deskripsi dari tiap kolom atau fitur, terdapat banyak kolom yang mengandung informasi dari masa depan.
Namun, karena tujuan kita adalah memprediksi apakah seseorang pelanggan akan membayar pinjaman atau tidak, maka model machine
learning yang kita gunakan harus ditraining hanya menggunakan data yang tersedia saat seseorang pertama kali mengajukan pinjaman
'''
# Membuang kolom-kolom yang mengandung informasi masa depan, informasi yang terlalu banyak, dan kolom yang tidak terlalu 
# memberikan informasi seperti id, member_id(hanya sebuah identifikasi tiap baris)
drop_column = ['id','member_id','funded_amnt','funded_amnt_inv','sub_grade','int_rate','emp_title','issue_d','zip_code',
              'out_prncp','out_prncp_inv','total_pymnt','total_pymnt_inv','total_rec_prncp','total_rec_int','total_rec_late_fee',
              'recoveries','collection_recovery_fee','last_pymnt_d','last_pymnt_amnt','next_pymnt_d','url']
df = df.drop(drop_column, axis=1)
df.head()
```

```python
# Melihat nilai unik kategori untuk tiap kolomnya yang kurang dari 4:
for col in df.columns:
    if len(df[col].unique()) < 4:
        print(df[col].value_counts())
        print()
```

```python
# Membuang kolom yang memiliki kategori tidak seimbang dan yang hanya berisi satu kategori
df = df.drop(['pymnt_plan','policy_code','application_type'], axis=1)
df.head()
```

```python
# Mengecek kembali missing value tiap kolomnya
print(len(df))
print(df.isnull().sum())
```

```python
# Menghapus tiga kolom terakhir karena missing value yang banyak
df = df.drop(['tot_coll_amt','tot_cur_bal','total_rev_hi_lim'], axis=1)
df.isnull().sum()
```

```python
# Melihat tiap kategori pada kolom emp_length
df['emp_length'].value_counts()
```

```python
# Mengisi missing value pada kolom emp_length dengan kategori terbanyak
df['emp_length'] = df['emp_length'].fillna(df['emp_length'].value_counts().idxmax())
df.isnull().sum()
```

```python
# Menghapus nilai missing value yang tersisa
df = df.dropna()
df.isnull().sum()
```

```python
# Melihat kolom kategori
categorical_columns_df = df.select_dtypes(include=['object'])
print(categorical_columns_df.iloc[0])
```

```python
# Melihat nilai unik dari tiap kolom kategori
for col in categorical_columns_df:
    print(col, ' :')
    print(df[col].value_counts())
    print()
```

```python
# Menghapus kolom yang nilai uniknya terlalu banyak
df = df.drop(['earliest_cr_line','last_credit_pull_d','addr_state','title'],axis=1)
```

```python
# Melihat korelasi tiap kolom atau fitur
plt.figure(figsize=(25,15))
sns.heatmap(df.corr(), annot=True)
```

```python
# kolom loan_amnt dan installment memiliki korelasi yang tinggi maka kita dapat membuang salah satunya
df = df.drop('installment', axis = 1)
plt.figure(figsize=(25,15))
sns.heatmap(df.corr(), annot=True)
```

```python
# Melihat kolom loan_status(kolom target)
df['loan_status'].value_counts()
```

```python
'''
Karena tujuan kita adalah memprediksi resiko kredit atau apakah seseoran akan membayar pinjaman atau tidak, 
maka target kita hanya status yang sudah tetap dan tidak akan berubah seperti Fully Paid dan Charged Off
'''
# Ambil hanya baris yang kolom statusnya Fully Paid atau Charged Off
df = df[(df["loan_status"] == "Fully Paid") | (df["loan_status"] == "Charged Off")]
print(len(df))
df.head(10)
```

```python
fig, axs = plt.subplots(1,2,figsize=(14,7))

sns.countplot(x='loan_status',data=df,ax=axs[0])

axs[0].set_title("Banyak dari tiap loan status")

df.loan_status.value_counts().plot(x=None,y=None, kind='pie', ax=axs[1],autopct='%1.2f%%')

axs[1].set_title("Persentase tiap loan status")

plt.show()
```

```python
# Mengubah nilai kategori dari loan status menjadi numerik 1/0
df = pd.get_dummies(df, columns=['loan_status'], drop_first=True)
df.head(10)
```

```python
# Mengubah kolom kategorik nominal(urutan tidak pengaruh) menjadi numerik
nominal_columns = ["initial_list_status","home_ownership", "verification_status", "purpose", "term"]
df = pd.get_dummies(df, columns=nominal_columns, drop_first=True)
df.head(10)
```

```python
# Mengubah kolom kategorik ordinal(urutan berpengaruh) menjadi numerik
mapping_dict = {

"emp_length": {"10+ years": 10, "9 years": 9, "8 years": 8, "7 years": 7,"6 years": 6,"5 years": 5,"4 years": 4,"3 years": 3,

"2 years": 2,"1 year": 1,"< 1 year": 0,"n/a": 0},

"grade":{"A": 1,"B": 2,"C": 3,"D": 4,"E": 5,"F": 6,"G": 7}
}

df = df.replace(mapping_dict)
df.head(10)
```

```python
# Import model machine learning dan metric
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('filtered_loan_data_v2.csv')
```

```python
smote = SMOTE()
y = df['loan_status_Fully Paid']
X = df.drop('loan_status_Fully Paid',axis=1)
scaler = StandardScaler()  
X_OverSmote, Y_OverSmote = smote.fit_resample(X, y)
X_scale = scaler.fit_transform(X_OverSmote)
X_final = np.array(X_scale)
y_final = np.array(Y_OverSmote)
```

```python
# Membagi data menjadi train dan test
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, random_state=0, test_size=.1)
```

```python
# Melakukan training data
rf = RandomForestClassifier(max_depth=None, max_features=10, n_estimators=15, random_state=2,verbose=2)
rf.fit(X_train, y_train)
```

```python
# Melihat performa model
y_pred = rf.predict(X_test)

print(accuracy_score(y_test,y_pred))
print("\n")
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm, index= ['Charged Off','Fully Paid'], columns=['Charged Off','Fully Paid'])
plt.figure(figsize=(10,5))
sns.heatmap(cm_df, annot=True, fmt='d', annot_kws={'size':20}, cmap='coolwarm')
plt.title('Confusion Matrix',fontsize=20)
plt.ylabel('Actual Values',fontsize=20)
plt.xlabel('Predicted Values',fontsize=20)
```
